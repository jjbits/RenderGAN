```python
from __future__ import absolute_import, division, print_function, unicode_literals

try:
  # %tensorflow_version only exists in Colab.
  %tensorflow_version 2.x
except Exception:
  pass
import tensorflow as tf

import os
import time

from matplotlib import pyplot as plt
from IPython import display
from matplotlib.pyplot import figure
from scipy import ndimage
import numpy as np
```

```python
def load(image_file):
  image = tf.io.read_file(image_file)
  image = tf.image.decode_png(image)

  return image
```

```python
PATH = "/Volumes/Data/projects/cs230/Project/pix2pix/pytorch-CycleGAN-and-pix2pix/checkpoints/plane995_pix2pix/web/images/"


figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')

inp = load(PATH+'epoch196_real_A.png')
plt.subplot(2, 3, 1)
plt.imshow(inp)
plt.title('input')
#plt.axis('off')

otp = load(PATH+'epoch196_fake_B.png')
plt.subplot(2, 3, 2)
plt.imshow(otp)
plt.title('generator output')
#plt.axis('off')

rel = load(PATH+'epoch196_real_B.png')
plt.subplot(2, 3, 3)
plt.imshow(rel)
plt.title('target output')
#plt.axis('off')

inp1 = load(PATH+'epoch189_real_A.png')
plt.subplot(2, 3, 4)
plt.imshow(inp1)
plt.title('input')
#plt.axis('off')

otp1 = load(PATH+'epoch189_fake_B.png')
plt.subplot(2, 3, 5)
plt.imshow(otp1)
plt.title('generator output')
#plt.axis('off')

rel1 = load(PATH+'epoch189_real_B.png')
plt.subplot(2, 3, 6)
plt.imshow(rel1)
plt.title('target output')
#plt.axis('off')

plt.savefig('1st_exp.png')
plt.show()
```
