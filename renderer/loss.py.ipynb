```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
from scipy import signal

def parse(log_filename):
    print(log_filename)
    
    file = open(log_filename, 'r')
    
    lines = []
    gen_loss = []
    l1_loss = []
    sum_loss = []
    for line in file:
      w_tokens = line.split()
      if w_tokens[0] == '(epoch:':
        lines.append(line)
        g_num = float(w_tokens[9])
        l1_num = float(w_tokens[11])
        gen_loss.append(g_num)
        l1_loss.append(l1_num)
        sum_loss.append((g_num+l1_num)/2)
        
    return lines, gen_loss, l1_loss, sum_loss
            

def parse0(LOG_FILE_PATH, LOG_FILE_NAME):
    log_filename = LOG_FILE_PATH + LOG_FILE_NAME
    print(log_filename)
    entries = [entry.strip().split('\t') \
               for entry in open(log_filename, 'r')]
    print(entries)
    
    iter = []
    errD = []
    errG = []
    L_img = []
    measure = []
    M = []
    for entry in entries:
        iter.append(entry[0])
        errD.append(entry[1])
        errG.append(entry[2])
        L_img.append(entry[3])
        measure.append(entry[4])
        M.append(entry[5])
        
    errD = np.asarray(errD).astype(np.float)
    errG = np.asarray(errG).astype(np.float)
    iter = np.asarray(iter).astype(np.float)
    L_img = np.asarray(L_img).astype(np.float)
    measure = np.asarray(measure).astype(np.float)
    M = np.asarray(M).astype(np.float)
    
    errD = scipy.signal.medfilt(errD, 101)[:-51];
    errG = scipy.signal.medfilt(errG, 101)[:-51];
    measure = scipy.signal.medfilt(measure, 101)[:-51];
    
    return errD, errG, iter, L_img, M, measure
```

```python
LOG_FILE_PATH = '/Volumes/Data/projects/cs230/Project/pix2pix/pytorch-CycleGAN-and-pix2pix/checkpoints/10-10000/'
LOG_FILE_NAME = 'loss_log.txt'
lines, gen_loss, l1_loss, sum_loss = parse(LOG_FILE_PATH+LOG_FILE_NAME)

N = 5
res = sorted(range(len(gen_loss)), key = lambda sub: gen_loss[sub], reverse=True)[-N:]
print("Gen loss winers")
for i in res:
    print(lines[i])
    
res = sorted(range(len(l1_loss)), key = lambda sub: l1_loss[sub], reverse=True)[-N:]
print("L1 loss winers")
for i in res:
    print(lines[i])
    
res = sorted(range(len(sum_loss)), key = lambda sub: sum_loss[sub], reverse=True)[-N:]
print("Avg loss winers")
for i in res:
    print(lines[i])
```
